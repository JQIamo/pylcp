```python
import numpy as np
import numba
```

Generally, an vector observable has a shape (3, ...)

```python
A = np.arange(300.).reshape(3, 100)
print(A)
```

A field gradient has at a minimum a shape (3, 3)

```python
B = np.array([[-0.5, 0, 0],[0., -0.5, 0.],[0, 0, 1.]])
print(B)
```

But it could have a shape (3, 3, ...):

```python
C = np.repeat(B, 100).reshape(3, 3, 100)
print(C)
```

Current way it's done:

```python
def default_dot(A, C):
    f_mag = np.zeros(A.shape)
    for ii in range(3): # Loop over muxB_x, mu_yB_y, mu_zB_z
        f_mag += A[ii]*C[:, ii]
    return f_mag

f_mag = default_dot(A, C)
print(f_mag)
```

Can we do it more compactly?

```python
np.sum(A[np.newaxis, ...]*C[...], axis=1)
```

```python
timeit np.sum(A[np.newaxis, ...]*C[...], axis=1)
```

```python
np.sum(A[np.newaxis, ...]*B[..., np.newaxis], axis=1)
```

```python
timeit np.sum(A[np.newaxis, ...]*B[..., np.newaxis], axis=1)
```

```python
timeit default_dot(A, C)
```

Now let's test performance on a single point:

```python
a = np.arange(3,)
print(default_dot(a, B))
```

```python
timeit default_dot(a, B)
```

Can we write a better performing function?

```python
def cartesian_vector_tensor_dot(a, B):
    if B.ndim == 2 and a.ndim == 1:
        # Single point:
        return np.dot(B, a)
    elif B.ndim == 2:
        # Constant B, variable a:
        return np.sum(a[np.newaxis, ...]*B[..., np.newaxis], axis=1)
    else:
        # Varaible a and variable B:
        return np.sum(a[np.newaxis, ...]*B[...], axis=1)
```

```python
cartesian_vector_tensor_dot(A, B)
```

```python
timeit cartesian_vector_tensor_dot(A, B)
```

```python
timeit cartesian_vector_tensor_dot(A, C)
```

```python
timeit cartesian_vector_tensor_dot(a, B)
```

```python
cartesian_vector_tensor_dot(a, B)
```
